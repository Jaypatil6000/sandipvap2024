```python
df =pd.read_csv('student.csv')
```

```python
plt.title('Roll vs. Marks')
plt.xlabel('Roll Number')
plt.ylabel('Marks')
plt.grid()
plt.xlim(0,10)
plt.ylim(50,100)
plt.plot(df['roll'] , df['marks'], color ='black', marker = '>',mfc='red' )
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xlim(0,70)
plt.ylim(0,100)
plt.yticks(range(0,100,10))
plt.plot(x, y, color ='brown', label ='2023')
plt.plot(x,z , color ='yellow', label ='2024')
plt.legend()
```
