# Normalization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df
```

```python
df.describe()
```

```python
x =df[['Age','EstimatedSalary']]
```

```python
x
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(df['Age'],df['EstimatedSalary'], c=df['Purchased'])
```

```python
import seaborn as sns
sns.jointplot(x='Age', y='EstimatedSalary', data=df , hue='Purchased')
```

```python
# Import the class
from sklearn.preprocessing import MinMaxScaler
```

```python
# Create the object
scaler = MinMaxScaler()
```

```python
# Apply the method
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
```

```python
x_scaled
```

```python
x_scaled.describe()
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x_scaled['Age'],x_scaled['EstimatedSalary'], c=df['Purchased'])
```
